O Population Stability Index (PSI) é uma métrica utilizada para avaliar se a distribuição de uma variável mudou ao longo do tempo. Em modelos de crédito, ele é fundamental para garantir que o comportamento da população atual seja semelhante ao observado durante o treinamento do modelo. Valores baixos de PSI indicam estabilidade, enquanto valores elevados sugerem risco de degradação do modelo.
